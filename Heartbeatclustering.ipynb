```python
import wave
import numpy as np
import os

# Ambil semua file dalam direktori set_b
files = os.listdir('./set_b')
total_duration = 0

# Loop untuk setiap file audio
for file in files:
    if not file.endswith('.wav'):
        continue  # Lewati file yang bukan WAV

    print(f"Processing file: {file}")

    # Buka file WAV
    with wave.open('./set_b/' + file, 'rb') as wav_file:
        num_channels = wav_file.getnchannels()
        sample_width = wav_file.getsampwidth()
        frame_rate = wav_file.getframerate()  # Sample rate (samples per second)
        num_frames = wav_file.getnframes()
        duration = num_frames / float(frame_rate)  # dalam detik

        print(f"Channels: {num_channels}")
        print(f"Sample Width: {sample_width} bytes")
        print(f"Frame Rate: {frame_rate} Hz")
        print(f"Number of Frames: {num_frames}")
        print(f"Duration: {duration:.2f} seconds")

        # Tambahkan durasi file ini ke total durasi
        total_duration += duration

print(f"Total Duration of All Audio Files: {total_duration:.2f} seconds")


```

```python
frames = wav_file.readframes(num_frames)
wav_file.close()                          

audio_data = np.frombuffer(frames, dtype=np.int16)

if num_channels > 1:
    audio_data = np.reshape(audio_data, (-1, num_channels))

print("Audio Array:", audio_data)
print("Array Shape:", audio_data.shape)

plt.figure(figsize=(10, 4))
if num_channels > 1: 
    plt.plot(audio_data[:, 0])  
else:
    plt.plot(audio_data)        
plt.title('Waveform of the Audio Signal')
plt.xlabel('Sample Number')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
import numpy as np                   # Untuk manipulasi array
import pandas as pd                  # Untuk manipulasi DataFrame (jika diperlukan)
import matplotlib.pyplot as plt       # Untuk visualisasi
from sklearn.preprocessing import StandardScaler  # Untuk standarisasi fitur
from sklearn.cluster import KMeans    # Untuk clustering dengan KMeans
from sklearn.metrics import silhouette_score  # Untuk menghitung Silhouette Score
import umap                           # Untuk reduksi dimensi dengan UMAP


# Ubah audio menjadi data berbasis fitur untuk clustering
# Standarisasi fitur menggunakan StandardScaler
scaler = StandardScaler()
audio_data_scaled = scaler.fit_transform(audio_data.reshape(-1, 1))

# Menggunakan UMAP untuk reduksi dimensi
umap_reducer = umap.UMAP(n_components=2, random_state=42)
audio_umap = umap_reducer.fit_transform(audio_data_scaled)
```

```python
# Clustering dengan KMeans
n_clusters = 3  # Bisa disesuaikan
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
y_kmeans = kmeans.fit_predict(audio_umap)

# Visualisasi hasil clustering
plt.figure(figsize=(10, 8))
plt.scatter(audio_umap[:, 0], audio_umap[:, 1], c=y_kmeans, cmap='Spectral', s=10, label='Data points')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='black', s=300, alpha=0.6, label='Cluster centers')
plt.title('UMAP Projection with KMeans Cluster Centers')
plt.legend()
plt.show()

# Evaluasi KMeans
silhouette_kmeans = silhouette_score(audio_umap, y_kmeans)
print(f"KMeans Silhouette Score: {silhouette_kmeans:.4f}")
```

```python
from sklearn.cluster import DBSCAN

# Clustering dengan DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(audio_umap)

# Visualisasi hasil clustering
plt.figure(figsize=(10, 8))
plt.scatter(audio_umap[:, 0], audio_umap[:, 1], c=y_dbscan, cmap='Spectral', s=10)
plt.title('UMAP Projection with DBSCAN Clustering')
plt.show()

# Evaluasi DBSCAN (jika lebih dari 1 cluster terbentuk)
if len(set(y_dbscan)) > 1:
    silhouette_dbscan = silhouette_score(audio_umap, y_dbscan)
    print(f"DBSCAN Silhouette Score: {silhouette_dbscan:.4f}")
else:
    print("DBSCAN tidak membentuk lebih dari 1 cluster.")
```

```python
from sklearn.cluster import MeanShift

# Clustering dengan MeanShift
meanshift = MeanShift()
y_meanshift = meanshift.fit_predict(audio_umap)

# Visualisasi hasil clustering
plt.figure(figsize=(10, 8))
plt.scatter(audio_umap[:, 0], audio_umap[:, 1], c=y_meanshift, cmap='Spectral', s=10)
plt.title('UMAP Projection with MeanShift Clustering')
plt.show()

# Evaluasi MeanShift
silhouette_meanshift = silhouette_score(audio_umap, y_meanshift)
print(f"MeanShift Silhouette Score: {silhouette_meanshift:.4f}")
```
